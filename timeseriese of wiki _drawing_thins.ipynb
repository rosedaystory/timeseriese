```python
def draw(start,num):
    """
    draw the timeseriese from start(row number) to number of num
    start = start point
    num = number of you want to draw
    """
    
    k = start
    plt.figure(figsize = (10,2*(num)))
    for i in range(num):
        plt.subplot(num,1,i+1)
        plt.plot(range(len(TS.iloc[k+i,1:])),TS.iloc[k+i,1:])
        plt.ylim(0,100)
    plt.show()
    
def draw_ylim(start,num,lim):
    """
    draw the timeseriese from start(row number) to number of num
    start = start point
    num = number of you want to draw
    """
    
    k = start
    plt.figure(figsize = (10,2*(num)))
    for i in range(num):
        plt.subplot(num,1,i+1)
        plt.plot(range(len(TS.iloc[k+i,1:])),TS.iloc[k+i,1:])
        plt.ylim(0,lim)
    plt.show()
```

```python
def draw_el(list):
    """
    draw the timeseriese in the list (each list)
    list = list that contain number of timeserieses that you want to draw!!
    """
    
    plt.figure(figsize = (10,2*(len(list))))
    for i in range(len(list)):
        plt.subplot(len(list),1,i+1)
        plt.plot(range(len(TS.iloc[list[i],1:])),TS.iloc[list[i],1:])
        plt.ylim(0,100)
    plt.show()
```
